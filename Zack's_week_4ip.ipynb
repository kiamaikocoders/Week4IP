**1.5.1 Validity**


```python
# Procedure 1: Irrelevant Observation
# Data Cleaning Action: Dropping columns 
# Explanation:We do not need them or they do not in fit the context of the problem

Electric.drop(['Displayed comment', 'ID', 'Geo point','Address', 'Cars'], axis = 1, inplace = True)
Electric
```

```python
# Procedure 2: Outliers
# Data Cleaning Action: checking for outliers using IQR Scores
# Explanation:


Q1 = Electric.quantile(0.25)
Q3 = Electric.quantile(0.75)

IQR = Q3 - Q1


(Electric < (Q1 - 1.5 * IQR)) |(Electric > (Q3 + 1.5 * IQR))


# The two commented lines of code below can be used to drop the outliers but for now I will not drop them

Electric_no_outlier = Electric[~((Electric < (Q1 - 1.5 * IQR)) | (Electric > (Q3 + 1.5 * IQR))).any(axis = 1)]

Electric_no_outlier
```

**1.6 Answering Research Questions**

1.6.1 Bluecar counter

```python
# In this week's independent project, I will be acting as a data scientist working for an electric car-sharing service company. 

# I have been tasked to process stations data to understand electric car usage over time by solving for the following research question;

# Research Question

# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018


# Do the results change if you consider Utilib and Utilib 1.4 instead of Bluecars? 
```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

# positive value mean a car was returned

# Negative value means a car was taken

# select city == Paris

Electric_paris = Electric2[Electric2['City'] =='Paris'] 

Electric_paris.head()
```

```python
# Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018

Electric_paris[Electric_paris['Bluecar_Diff'] < 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python
# Q1

# What is the most popular hour for returning cars?

Electric_paris[Electric_paris['Bluecar_Diff'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python
# Q2a

# What station is the most popular?

  # Overall?

Electric_paris[(Electric_paris['Kind'] == 'STATION') & (Electric_paris['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending= False)
```

```python
# Q2b

# What station is the most popular? At the most popular picking hour?
  
Electric_paris[(Electric_paris['Kind'] == 'STATION') & (Electric_paris['Status'] == 'ok') & (Electric_paris['hour'] == 21)].groupby('Public name')['Public name'].count().sort_values(ascending= False)
```

```python
# Q3a

#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?

 # Overall?

Electric_paris[(Electric_paris['Kind'] == 'STATION') & (Electric_paris['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)

    #from the dataframe 75015 == Paris/Houdart de Lamotte/2  #therefore NO it doesn't
```

```python
# Q3b

#  What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?

# At the most popular picking hour?

Electric_paris[(Electric_paris['Kind'] == 'STATION') & (Electric_paris['Status'] == 'ok') & (Electric_paris['hour'] == 21)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)
```
